# Inspect en Valideer 'Book1 (1).xlsx'
Dit notebook wordt gebruikt om de inhoud van het Excel-bestand `Tijdelijke Bestanden/Book1 (1).xlsx` te inspecteren en te valideren.

## 1. Importeer Vereiste Bibliotheken
We importeren `pandas` voor data-analyse en `os` voor padbeheer.

```python
import pandas as pd
import os

print("Bibliotheken geladen.")
```

## 2. Definieer Bestandspad en Controleer Bestaan
We definiëren het pad naar het bestand en controleren of het toegankelijk is.

```python
file_path = 'Tijdelijke Bestanden/Book1 (1).xlsx'
if os.path.exists(file_path):
    print(f"Bestand gevonden: {file_path}")
else:
    print(f"FOUT: Bestand niet gevonden op {file_path}")
```

## 3. Laad Excel Bestand in DataFrame
We laden de gegevens met de `openpyxl` engine. (We proberen de benodigde packages te installeren indien ze ontbreken).

```python
try:
    df = pd.read_excel(file_path, engine='openpyxl')
    print("Gegevens succesvol geladen.")
except Exception as e:
    print(f"Fout bij laden: {e}")
    print("Probeer 'pip install pandas openpyxl' in de terminal.")
```

## 4. Inspecteer Datastructuur en Typen
We bekijken de kolommen en de eerste rijen.

```python
print("--- Info ---")
print(df.info())
print("\n--- Eerste 5 rijen ---")
display(df.head())
```

## 5. Controleer op Ontbrekende Waarden
We tellen het aantal null-waarden per kolom.

```python
print("--- Ontbrekende waarden ---")
print(df.isnull().sum())
```

## 6. Genereer Samenvattende Statistieken
Statistische metrieken zoals gemiddelde $\mu$ en standaardafwijking $\sigma$ voor numerieke kolommen.

```python
print("--- Statistieken ---")
display(df.describe())
```

## 7. Gegevensvalidatie en Kwaliteitscontroles
We voeren specifieke controles uit, zoals het controleren op unieke waarden.

```python
# Voorbeeld: Controleer of de eerste kolom uniek is (vaak een ID of Lotnummer)
if not df.empty:
    first_col = df.columns[0]
    is_unique = df[first_col].is_unique
    print(f"Is de kolom '{first_col}' uniek? {is_unique}")
    if not is_unique:
        print(f"Aantal duplicaten: {df[first_col].duplicated().sum()}")
```

```python
# Toon de eerste 20 rijen van relevante kolommen (Unnamed: 1 t/m Unnamed: 16)
display(df.iloc[:20, 1:17])
```

```python
# Print de waarden van de eerste 10 rijen voor een subset van de kolommen om de structuur te begrijpen
print(df.iloc[:10, 1:10].to_string())
```

```python
# Print slechts de eerste 3 rijen om de headers/inhoud te zien
for i in range(3):
    print(f"Rij {i}: {df.iloc[i, 1:17].tolist()}")
```

```python
# Zoek de eerste rij die NIET alleen uit NaN bestaat
non_empty_rows = df.dropna(how='all', subset=df.columns[1:17])
if not non_empty_rows.empty:
    print(f"Eerste niet-lege rij index: {non_empty_rows.index[0]}")
    print(f"Inhoud: {non_empty_rows.iloc[0, 1:17].tolist()}")
else:
    print("Geen niet-lege rijen gevonden in de opgegeven kolommen.")
```

```python
# Toon rijen 4 t/m 15 om de tabelstructuur te zien
display(df.iloc[4:15, 1:17])
```

```python
# Print rijen 4 t/m 10 om de inhoud te zien zonder dat het wordt afgebroken
for i in range(4, 11):
    print(f"Rij {i}: {df.iloc[i, 1:17].tolist()}")
```
